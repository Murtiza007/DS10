```python
import pandas as pd
df=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
df.head()
```

```python
df.shape
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
transaction=[]
for i in range(0,len(df)):
    transaction.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0' ])
```

```python
transaction[4]
```

```python
#!pip install apyori
from apyori import apriori
rules=apriori(transactions=transaction,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=4)
```

```python
Result=list(rules)
len(Result)
```

```python
Result[1]
```

```python
for items in Result:
    pair=items[0]
    for itemss in pair:
        print(itemss)

    support=items[1]
    confidence=items[2][0][2]
    lift=items[2][0][3]

    print("Support = ",support)
    print("Confidence = ",confidence)
    print("Lift = ",lift)
    print("----------------------------------------------------------------------------------------------------------------")
```
