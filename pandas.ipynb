```python
import pandas as pd


tiles={
    'floor length':[20,30,40,50,60],
    'floor breadth':[10,30,30,20,70],
    'Tile lenght':[3,2,4,5,6],
    'Tile breadth':[1,2,4,2,5]   
}
```

```python
df=pd.DataFrame(tiles)
df
```

```python
df['floor area']=df['floor length']*df['floor breadth']
df
```

```python
df['tile area']= df['Tile lenght']*df['Tile breadth']
df
```

```python
df['no of tile']=df['floor area']/df['tile area']
df
```
