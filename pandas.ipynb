```python
marks_1={'Roll No':[1,2,3,4,5],
         'Names':['Rameez','Showkat','Raheel','Aabid','Sadiya'],
         'Science_marks':[50,60,40,50,50]}
```

```python
marks_2={'Roll No':[1,2,3,4,5],
         'Names':['Rameez','Showkat','Raheel','Aabid','Sadiya'],
         'Social_marks':[40,50,50,30,60]}
```

```python
import pandas as pd
df1=pd.DataFrame(marks_1)
df1
```

```python
df2=pd.DataFrame(marks_2)
df2
```

```python
df3=pd.merge(df1,df2,on=['Roll No','Roll No'])
df3
```
