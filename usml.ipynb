```python
import pandas as pd
data=pd.read_csv('income.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Age'],data['Income($)'])
plt.xlabel('age')   
plt.ylabel('income')
plt.show()
```

```python
from sklearn.cluster import KMeans

sse=[]
k_range=range(2,8)
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(data[['Age','Income($)']])
    sse.append(km.inertia_)
    
    
plt.plot(k_range,sse)
# sse
```

```python
km=KMeans(n_clusters=5)
y_pred=km.fit_predict(data[['Age','Income($)']])
 
```

```python
data['cluster']=y_pred
data
```

```python
km.cluster_centers_
```

```python
df1=data[data['cluster']==0]
df2=data[data['cluster']==1]
df3=data[data['cluster']==2]
df4=data[data['cluster']==3]
df5=data[data['cluster']==4]

plt.scatter(df1['Age'],df1['Income($)'],color='red')
plt.scatter(df2['Age'],df2['Income($)'],color='blue')
plt.scatter(df3['Age'],df3['Income($)'],color='green')
plt.scatter(df4['Age'],df4['Income($)'],color='orange')
plt.scatter(df5['Age'],df5['Income($)'],color='purple')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker='*',s=200)

plt.xlabel('age')
plt.ylabel('income')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('Mall_customers.csv')
data.head()
```

```python
plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
df=data.iloc[:,3:5].values
df.shape
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(df,method='ward'))
plt.title('Dendrogram')
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,linkage='ward')
labels=hc.fit_predict(df)
```

```python
data['cluster']=labels
```

```python
data
```

```python
df1=data[data['cluster']==0]
df2=data[data['cluster']==1]
df3=data[data['cluster']==2]
df4=data[data['cluster']==3]
df5=data[data['cluster']==4]

plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color='red')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color='blue')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color='green')
plt.scatter(df4['Annual Income (k$)'],df4['Spending Score (1-100)'],color='orange')
plt.scatter(df5['Annual Income (k$)'],df5['Spending Score (1-100)'],color='purple')
```
